```python
import numpy as np
import os
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pickle
import cv2

# Load the dataset
data_path = "C:/Users/Mariam Usman/Vehicle_dataset"
categories = os.listdir(data_path)
labels = [i for i in range(len(categories))]

# Define a function to extract features from an image
def extract_features(image_path):
    # load the image from the given file path
    img = cv2.imread(image_path, cv2.IMREAD_ANYCOLOR)
    # resize the image to a fixed size
    img = cv2.resize(img, (128, 128))
    # flatten the image to a 1D numpy array
    feature = img.flatten()
    return feature.astype(np.float32)

# Extract features and labels from the dataset
data = []
for category in categories:
    folder_path = os.path.join(data_path, category)
    image_names = os.listdir(folder_path)
    for image_name in image_names:
        image_path = os.path.join(folder_path, image_name)
        if image_path.endswith(('.jpg', '.jpeg', '.png')):    
            feature = extract_features(image_path)
            label = labels[categories.index(category)]
            data.append((feature, label))
        
# Split the data into training and testing sets
features, labels = zip(*data)
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.1, random_state=72)

# Train an SVM model
clf = svm.SVC(kernel='rbf', C=9, probability=True)
clf.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))

# Save the SVM model to a file
# Save the model to a file using pickle
with open('model1.pkl', 'wb') as f:
    pickle.dump(clf, f)
```

```python
import cv2
import numpy as np
import tkinter as tk
from tkinter import filedialog
import PIL.Image, PIL.ImageTk
import os
import pickle

# Load the SVM model from a file
with open('model1.pkl', 'rb') as f:
    clf = pickle.load(f)

# Define a function to extract features from an image
def extract_features(image_path):
    # load the image from the given file path
    img = cv2.imread(image_path)
    # resize the image to a fixed size
    img = cv2.resize(img, (128, 128))
    # flatten the image to a 1D numpy array
    feature = img.flatten()
    return feature.astype(np.float32)

# Define a function to choose an image file and display it along with the predicted label
def choose_image_and_predict_label():
    # Choose an image file
    file_path = filedialog.askopenfilename()
    
    # Load the image from the file
    img = cv2.imread(file_path)
    
    # Resize the image and convert it to a format that can be displayed in Tkinter
    img = cv2.resize(img, (400, 400))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = PIL.Image.fromarray(img)
    img_tk = PIL.ImageTk.PhotoImage(img)
    
    # Display the image in a label widget
    img_label.configure(image=img_tk)
    img_label.image = img_tk
    
    # Extract features from the image and make a prediction using the SVM model
    feature = extract_features(file_path)
    label_names = ['bicycle', 'Bike', 'Car']
    label_idx = clf.predict(feature.reshape(1,-1))[0]
    predicted_label = label_names[label_idx]
    
    # Display the predicted label in a label widget
    label_label.configure(text=f"Predicted label: {predicted_label}")

# Create a Tkinter window
root = tk.Tk()

# Create a label widget to display the image
img_label = tk.Label(root)
img_label.pack()

# Create a label widget to display the predicted label
label_label = tk.Label(root, font=("Helvetica", 16))
label_label.pack()

# Create a button to choose an image file and display it
button = tk.Button(root, text="Choose Image", command=choose_image_and_predict_label)
button.pack()

# Start the Tkinter event loop
root.mainloop()
```
